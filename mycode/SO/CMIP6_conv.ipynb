```python
from intake import open_catalog
url = "https://raw.githubusercontent.com/pangeo-data/pangeo-datastore/master/intake-catalogs/master.yaml"
col = open_catalog(url).climate.cmip6_gcs()
```

```python
modellist = col.search(variable_id = ['so', 'thetao'], table_id = 'Omon', 
                       experiment_id = 'piControl').df.source_id.unique()
print(len(modellist))
modellist
```

```python
modellist.sort()    
modellist
```

```python
newmodelid = []
modeldict = {}
for modelid in modellist:
    dslist = col.search(variable_id = ['so', 'thetao'], table_id = 'Omon', 
                        source_id = modelid, experiment_id = 'piControl', grid_label = 'gn')
    if dslist:
        model0 = dslist[list(dslist)[0]]
        if len(model0.df['variable_id']) >= 2:
            try:
                modeldict[modelid] = dslist[list(dslist)[0]].to_dask()
                print(modelid)
            except:
                newmodelid.append(modelid)
        else:
            newmodelid.append(modelid)        
    else:
        newmodelid.append(modelid)
```
